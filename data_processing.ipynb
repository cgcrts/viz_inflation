```python
import pandas as pd
```

```python
old_file_date = '10_15_it'
new_file_date = '11_01'
new_date_col = '01.11.2022'

with open('data/inflation_data_updated_' + old_file_date +'.csv', encoding='utf-8') as file:
    data_df = pd.read_csv(file, index_col=0)
    
data_df
```

```python
with open('data/inflation_data_' + new_file_date + '.csv', encoding='utf-8') as file:
    recent_data_df = pd.read_csv(file)
    
recent_data_df
```

```python
updated_df = data_df.copy()
counter = 0

for idx, row in data_df.iterrows():
    url = row.url
    shop = row.shop
    product_id = row.product_id
    
    recent_matches = recent_data_df.loc[recent_data_df['URL'] == url]
    
    if len(recent_matches) == 1:
        recent_row = recent_matches.iloc[0]
        new_price = recent_row[new_date_col]
        
        updated_df.loc[idx, new_date_col] = new_price
        counter += 1
    elif len(recent_matches) > 1:
        print(recent_matches)
    else:
        print('-------- no match --------', row.product_id)
        
print(counter)
updated_df
```
