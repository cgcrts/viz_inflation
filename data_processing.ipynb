```python
import pandas as pd
```

```python
old_file_date = '05_18'
new_file_date = '06_05'
new_date_col = '05.06.2023'
new_filename = 'inflation_data_updated_' + new_file_date

with open('data/inflation_data_updated_' + old_file_date +'.csv', encoding='utf-8') as file:
    old_data_df = pd.read_csv(file, index_col=0)
    
old_data_df
```

```python
with open('data/inflation_data_' + new_file_date + '.csv', encoding='utf-8') as file:
    new_data_df = pd.read_csv(file)
    new_data_df = new_data_df.dropna(subset=['URL'])
    new_data_df.reset_index(drop=True, inplace=True)
    
new_data_df
```

```python
# get product not found in new_data

new_url_list = list(new_data_df['URL'])
old_url_list = list(old_data_df['url'])

old_idx_to_drop = list()
for idx, row in old_data_df.iterrows():
    old_url = row['url']
    if old_url not in new_url_list:
        print(old_url)
        old_idx_to_drop.append(idx)

old_data_df.iloc[old_idx_to_drop]
```

```python
# get product not found in old data

new_idx_to_drop = list()
for idx, row in new_data_df.iterrows():
    new_url = row['URL']
    if new_url not in old_url_list:
        new_idx_to_drop.append(idx)

new_data_df.iloc[new_idx_to_drop]
```

```python
"""
Produits manquants:
- Disques de ouate ronde, Coop: déjà des disques de ouate coop
- Chips Pringles, Migros: déjà des chips Migros
- Gel douche, Migros: déjà du gel douche Migros
- Papier toilette, Migros: déjà du pq Migros
- Yogourt vanille, Coop: déjà yogourt coop et yogourt vanille migros
"""
```

```python
updated_df = old_data_df.copy()
counter = 0

for idx, row in old_data_df.iterrows():
    url = row.url
    shop = row.shop
    product_id = row.product_id
    
    new_matches = new_data_df.loc[new_data_df['URL'] == url]
    
    if len(new_matches) == 1:
        new_row = new_matches.iloc[0]
        new_price = new_row[new_date_col]
        
        updated_df.loc[idx, new_date_col] = new_price
        counter += 1
    elif len(new_matches) > 1:
        print(new_matches)
    else:
        print('-------- no match --------', row.product_id)
        
print(counter)
updated_df
```
